```python
def confusionMatrix(y_true, y_pred):
    target_names = ['Non-binder', 'Binder']
    print(classification_report(y_true, y_pred, target_names=target_names))
    cm = pd.DataFrame(confusion_matrix(y_true, y_pred), index=target_names, columns=target_names)
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='g')
    plt.title('Confusion matrix')
    plt.xlabel('Predicted values')
    plt.ylabel('Actual values')
    plt.show()
    
def data_visu(data):
    y_test = data["binder"].to_numpy()
    y_test_pred = data["binder_pred"].to_numpy()

    confusionMatrix(y_test, y_test_pred)

    tn, fp, fn, tp = confusion_matrix(y_test, y_test_pred).ravel()
    accuracy = float(accuracy_score(y_test, y_test_pred))
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    auc = roc_auc_score(y_test, data["prediction_score"])    
    
    print("AUC: ", round(auc, 3))
    print("Accuracy score: ", round(accuracy, 3))
    print('Sensitivity (TPR): ', round(sensitivity, 3))
    print('Specificity (TNR): ', round(specificity, 3))
    
# data_visu(data_imrex)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

def fn_split_samplingdata(df):
    num_observations = len(df)
    num_samples = 10
    sample_size = int(num_observations / num_samples)

    indices = np.random.permutation(num_observations)

    samples = [indices[i * sample_size:(i + 1) * sample_size] for i in range(num_samples)]

    sampled_dataframes = [df.iloc[sample] for sample in samples]
    
    return sampled_dataframes
```

```python
sampled_dataframes_KD       = fn_split_samplingdata(KD_Overral)
sampled_dataframes_epitcr   = fn_split_samplingdata(epiTCR_Overral)
sampled_dataframes_nettcr   = fn_split_samplingdata(NetTCR_Overral)
sampled_dataframes_bertrand = fn_split_samplingdata(Bertrand_Overral)
sampled_dataframes_teim     = fn_split_samplingdata(TEIM_Overral)
sampled_dataframes_teinet   = fn_split_samplingdata(TEInet_Overral)
sampled_dataframes_imrex    = fn_split_samplingdata(Imrex_Overral)
```
